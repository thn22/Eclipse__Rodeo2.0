```python
import pandas as pd
import numpy as np
```

```python
#file path
netflix_credits = "Resources/credits(Netflix).csv"
netflix_title = "Resources/titles(Netflix).csv"
hulu_credits = "Resources/credits(Hulu).csv"
hulu_title = "Resources/titles(Hulu).csv"
hbo_credits = "Resources/credits(HBO).csv"
hbo_title = "Resources/titles(HBO).csv"
```

```python
#read in csv files
net_credit_df = pd.read_csv(netflix_credits)
net_title_df = pd.read_csv(netflix_title)
hulu_credit_df = pd.read_csv(hulu_credits)
hulu_title_df = pd.read_csv(hulu_title)
hbo_credit_df = pd.read_csv(hbo_credits)
hbo_title_df = pd.read_csv(hbo_title)
```

```python
#merge data frames
#merge Netflix first
netflix_df = pd.merge(net_credit_df, net_title_df, how='outer', on=['id','id'])
netflix_df['Netflix'] = 1

#merge Hulu
hulu_df = pd.merge(hulu_credit_df, hulu_title_df, how='outer', on=['id','id'])
hulu_df["Hulu"] = 1

#merge HBO
hbo_df = pd.merge(hbo_credit_df, hbo_title_df, how='outer', on=['id','id'])
hbo_df["HBO"] = 1

#netflix_df
#hulu_df
#hbo_df
```

```python
#merge all data frames into one
combined_df = pd.merge(pd.merge(netflix_df, hulu_df, how='outer'), hbo_df, how='outer')
combined_df
```

```python
combined_df = combined_df[['id', 'Netflix', 'Hulu', 'HBO', 'title', 'type', 'age_certification', 'runtime', 'genres', 'name', 'character', 'role', 'release_year', 'seasons', 'imdb_score', 'imdb_votes', 'production_countries']]
combined_df
```

```python
combined_df.dropna(subset=['imdb_score','genres'])
```

```python
combined_df = combined_df[combined_df['imdb_score'] > 6.5]
combined_df
```

```python
combined_df = combined_df[combined_df['imdb_votes'] > 10000]
combined_df
```

```python
combined_df = combined_df.fillna({"age_certification": '',
                    "name": '--',
                    "character": '--',
                    "role": '--',
                    "Netflix": 0,
                    "Hulu": 0,
                    "HBO": 0,
                    "seasons": 0})
combined_df
```

```python
movie_df = combined_df[combined_df["type"]=="MOVIE"]
movie_df
```

```python
show_df = combined_df[combined_df["type"]=="SHOW"]
show_df
```

```python
#export csv's to resources folder
from pathlib import Path
filepath = Path("Resources/movie_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
movie_df.to_csv(filepath)

filepath2 = Path("Resources/show_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
show_df.to_csv(filepath2)
```
