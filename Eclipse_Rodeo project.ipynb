```python
# Import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Create file path for each csv
netflix_credits = "Resources/credits(Netflix).csv"
netflix_title = "Resources/titles(Netflix).csv"
hulu_credits = "Resources/credits(Hulu).csv"
hulu_title = "Resources/titles(Hulu).csv"
hbo_credits = "Resources/credits(HBO).csv"
hbo_title = "Resources/titles(HBO).csv"
```

```python
# Read in csv files
net_credit_df = pd.read_csv(netflix_credits)
net_title_df = pd.read_csv(netflix_title)
hulu_credit_df = pd.read_csv(hulu_credits)
hulu_title_df = pd.read_csv(hulu_title)
hbo_credit_df = pd.read_csv(hbo_credits)
hbo_title_df = pd.read_csv(hbo_title)
```

```python
# Merge data frames
# Merge Netflix first, create "Netflix" column and add value of 1 to use as boolean for if movie/show is available
# stream on Netflix, also drop N/A values for IMDb votes and drop Actors to reduce dataset
netflix_df = pd.merge(net_credit_df, net_title_df, how='outer', on=['id','id'])
netflix_df['Netflix'] = 1
netflix_df = netflix_df.dropna(subset=['imdb_votes'])
netflix_df = netflix_df.drop(netflix_df[netflix_df["role"]=="ACTOR"].index)

# Merge Hulu using reasoning above
hulu_df = pd.merge(hulu_credit_df, hulu_title_df, how='outer', on=['id','id'])
hulu_df["Hulu"] = 1
hulu_df = hulu_df.dropna(subset=['imdb_votes'])
# hulu_df = hulu_df.dropna(subset=['imdb_score', 'imdb_votes'])
hulu_df = hulu_df.drop(hulu_df[hulu_df["role"]=="ACTOR"].index)

# Merge HBO using reasoning above
hbo_df = pd.merge(hbo_credit_df, hbo_title_df, how='outer', on=['id','id'])
hbo_df["HBO"] = 1
hbo_df = hbo_df.dropna(subset=['imdb_votes'])
hbo_df = hbo_df.drop(hbo_df[hbo_df["role"]=="ACTOR"].index)

# Confirm each df appears to be merged correctly by previewing it
#netflix_df
#hulu_df
#hbo_df
```

```python
# Drop "imdb_votes" column from all dataframe
netflix_df = netflix_df.drop(["imdb_votes"], axis=1)
hulu_df = hulu_df.drop(["imdb_votes"], axis=1)
hbo_df = hbo_df.drop(['imdb_votes'], axis=1)
```

```python
# Merge all data frames into one in a nested merge
combined_df = pd.merge(pd.merge(netflix_df, hulu_df, how='outer'), hbo_df, how='outer')
```

```python
# Isolate Christopher Nolan movies that are duplicates and delete them. Add 1's for the duplicated streaming service
# so info is still present
inception = combined_df[combined_df["title"]=="Inception"]
dunkirk = combined_df[combined_df["title"]=="Dunkirk"]
#inception
#dunkirk
combined_df.iloc[186,20]=1
combined_df.iloc[1268,20]=1
combined_df = combined_df.drop(labels=[6809, 7729], axis=0)
combined_df.head()
```

```python
# Reorder the columns and drop any that are unnecessary 
combined_df = combined_df[['id', 'Netflix', 'Hulu', 'HBO', 'title', 'type', 'age_certification', 'runtime', \
                           'name', 'role', 'release_year', 'seasons', 'imdb_score']]
#combined_df
```

```python
# Fill out all the empty rows with dashes or zeros
combined_df = combined_df.fillna({"name": '--',
                    "role": '--',
                    "Netflix": 0,
                    "Hulu": 0,
                    "HBO": 0,
                    "seasons": 0})
combined_df
```

```python
# Separate the combined dataframe into movies and TV shows dataframe
# Filter the combined dataframe for all movies
movie_df = combined_df[combined_df["type"]=="MOVIE"]
movie_df.head()
```

```python
# Filter the combined dataframe for all tv shows
show_df = combined_df[combined_df["type"]=="SHOW"]
show_df.head()
```

```python
# Filter the dataframe for Netflix movies data and fill out the empty rows in Seasons column with zeros.
netflix_df = netflix_df[['id', 'title', 'type', 'age_certification', 'runtime', 'release_year', \
                         'seasons', 'imdb_score']]
netflix_df = netflix_df.fillna({"name": '--',
                                "role": '--',
                                "seasons": 0})
netflix_df = netflix_df.drop_duplicates()
netflix_df
```

```python
# Filter the dataframe for Hulu movies data and fill out the empty rows in Seasons column with zeros.
hulu_df = hulu_df[['id', 'title', 'type', 'age_certification', 'runtime', 'release_year', \
                   'seasons', 'imdb_score']]
hulu_df = hulu_df.fillna({"name": '--',
                    "character": '--',
                    "role": '--',
                    "seasons": 0})
hulu_df = hulu_df.drop_duplicates()
hulu_df
```

```python
# Filter the dataframe for HBO movies data and fill out the empty rows in Seasons column with zeros.
hbo_df = hbo_df[['id', 'title', 'type', 'age_certification', 'runtime', 'release_year', \
                 'seasons', 'imdb_score']]
hbo_df = hbo_df.fillna({"name": '--',
                    "character": '--',
                    "role": '--',
                    "seasons": 0})
hbo_df = hbo_df.drop_duplicates()
hbo_df
```

```python
# Export csv's to resources folder
filepath = Path("Output/movie_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
movie_df.to_csv(filepath)

filepath2 = Path("Output/show_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
show_df.to_csv(filepath2)

filepath3 = Path("Output/netflix_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
netflix_df.to_csv(filepath3)

filepath4 = Path("Output/hulu_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
hulu_df.to_csv(filepath4)

filepath5 = Path("Output/hbo_df.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
hbo_df.to_csv(filepath5)
```
